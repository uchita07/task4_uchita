```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
#checking the missing value
data.isnull().sum()
```

```python
#droping the records whose response is missing in data
data = data[~data.Selling_Price.isnull()].copy()
```

```python
#calculating the missing values in each column of data frame
data.isnull().sum()
```

Scatter Plot

```python
plt.scatter(data.Selling_Price,data.Present_Price)
plt.show()
```

Pair Plot

```python
sns.pairplot(data=data,vars=['Selling_Price','Present_Price','Kms_Driven'])
plt.show()
```

Histogram


```python
plt.hist(data['Selling_Price'])
```

Bar Chart:-

```python
data.groupby('Selling_Price')['Present_Price'].mean().plot.bar()
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.barplot(data['Selling_Price'], data['Present_Price'])
```

Line Plot:-


```python
plt.figure(figsize=(20,6))
sns.lineplot(data=data['Selling_Price'],linewidth=1.5, label='Selling_Price')
sns.lineplot(data=data['Present_Price'],linewidth=1.5, label='Present_Price')
```
